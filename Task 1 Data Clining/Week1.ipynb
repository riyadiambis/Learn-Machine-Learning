## Cleaning Data

```python
data_kotor = pd.DataFrame(data.isnull().sum(), columns=["Jumlah Null"])
data_kotor
```

```python
# Sample data yang memiliki nilai null
data[['LastPrice', 'MarketCap', 'MinutesFirstAdded', 'MinutesLastUpdated', 'HourlyFirstAdded', 'HourlyLastUpdated', 'DailyFirstAdded', 'DailyLastUpdated']].sample(5)
```

```python
# Mengatasi nilai null pada kolom LasPrice 
data['LastPrice'] = data['LastPrice'].fillna(0)
data['LastPrice'].isnull().sum()
```

```python
# Mengatasi nilai null pada kolom MarketCap
data['MarketCap'] = data['Shares'] * data['LastPrice']
data['MarketCap'].isnull().sum()
```

```python
# Menampilkan jumlah persentase untuk pertimbangan mengatasi sisa data null
pd.DataFrame((data.isnull().sum() / len(data) * 100), columns=["Persentasi Data Kosong"])
```

```python
# MinutesFirstAdded	(Waktu pertama kali data dicatat) seharusnya bisa diganti dengan ListingDate(Tanggal pertama kali saham dicatatkan di BEI.)
data['MinutesFirstAdded'] = data['MinutesLastUpdated'].fillna(data['ListingDate'])
data['MinutesFirstAdded'].isnull().sum()
```

```python
# Hal ini sama dengan MinutesLastUpdate (Waktu terakhir data diperbaharui)
data['MinutesLastUpdated'] = data['MinutesLastUpdated'].fillna(data['ListingDate'])
data['MinutesLastUpdated'].isnull().sum()
```

```python
data['HourlyFirstAdded'] = data['HourlyFirstAdded'].fillna(data['ListingDate'])
data['HourlyFirstAdded'].isnull().sum()
```

```python
data['HourlyLastUpdated'] = data['HourlyLastUpdated'].fillna(data['HourlyFirstAdded'])
data['HourlyLastUpdated'].isnull().sum()
```

```python
data['DailyFirstAdded'] = data['DailyFirstAdded'].fillna(data['ListingDate'])
data['DailyFirstAdded'].isnull().sum()
```

```python
tanggalMax = data['DailyLastUpdated'].max()
data['DailyLastUpdated'] = data['DailyLastUpdated'].fillna(data['DailyLastUpdated'])
data['DailyLastUpdated'].isnull().sum()
```

```python
# Menampilkan jumlah persentase untuk pertimbangan mengatasi sisa data null
pd.DataFrame((data.isnull().sum() / len(data) * 100), columns=["Persentasi Data Kosong"])
```

## Data Cleaning

```python
def remove_outliers_iqr(df, columns, x):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean
```

```python
dataBersih = remove_outliers_iqr(data, dataNumerik, 2)
dataBersih
```
